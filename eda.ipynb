# У женщин более высокая устойчивость к смартфон-зависимости

```python
import seaborn as sns

# Настраиваем стиль и палитру заранее
sns.set_theme(style="whitegrid")
sns.set_palette("colorblind")  # научная палитра, оптимальна для дальтоников

g = sns.catplot(
    data=df,
    x="Dem_St_Sex",
    y="SAS-SV",
    kind="violin",
    order=[1, 2]  # задаем порядок категорий
).set(ylim=(0, 60))

# Шаг значений и подписи
for ax in g.axes.flat:
    ax.set_yticks(range(0, 61, 5))
    ax.set_xlabel("Пол", fontsize=12)
    ax.set_ylabel("SAS-оценка", fontsize=12)
    ax.set_title("Распределение зависимости от смартфона по полу", fontsize=14, pad=15)
    
    # Переименовываем подписи на оси X
    ax.set_xticklabels(['Мужской', 'Женский'])
```

```python
# Берём все SNP-столбцы, кроме SNP2 (без значений)
snp_cols = [col for col in df.columns if col.startswith("SNP") and not col.startswith("SNP2")]

# Выбираем числовые колонки для корреляции
num_cols = ['SAS-SV'] + snp_cols

# Строим корреляционную матрицу
corr = df[num_cols].corr()

# Настраиваем фигуру для читабельности
plt.figure(figsize=(15, 12))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Корреляции SAS-SV и генотипы (без SNP2)", fontsize=14)
plt.show()
```

```python
import math

# Настраиваем стиль и палитру
sns.set_theme(style="whitegrid", palette="colorblind")

# Список SNP для визуализации
snps = ['SNP1_174868700', 'SNP3_78172260', 'SNP4_18047255', 'SNP5_18047816', 'SNP6_113346955',
        'SNP7_113803028', 'SNP8_72331923', 'SNP9_72372862', 'SNP10_47409034', 'SNP11_28551665',
        'SNP12_19951271', 'SNP13_27677041', 'SNP14_27679916', 'SNP15_63261329', 'SNP16_47471478']

# Размер сетки
cols = 4
rows = math.ceil(len(snps)/cols)

# Создаем фигуру и оси
fig, axes = plt.subplots(rows, cols, figsize=(5*cols, 5*rows), sharey=True)

# Преобразуем axes в плоский массив для итерации
axes_flat = axes.ravel()

# Рисуем boxplots
for i, snp in enumerate(snps):
    sns.boxplot(data=df, x=snp, y='SAS-SV', ax=axes_flat[i])
    axes_flat[i].set_xlabel(snp)
    axes_flat[i].set_ylabel('SAS-оценка')
    axes_flat[i].set_title(f'SAS-SV vs {snp}')
    axes_flat[i].grid(True, linestyle='--', alpha=0.5)

# Убираем пустые графики, если они есть
for i in range(len(snps), len(axes_flat)):
    fig.delaxes(axes_flat[i])

plt.tight_layout()
plt.show()
```

Не все SNP одинаково влияют на SAS-SV, что соответствует полигенной природе поведенческих признаков. Разные SNP вносят различный вклад в фенотип
 
 У некоторых SNP наблюдается четкий сдвиг медианных значений между разными генотипами, что свидетельствует о систематическом влиянии генетического варианта на выраженность признака. Наблюдаемые сдвиги медиан указывают на аллель-специфические эффекты
  
  SNP 1 будем рассматривать как наиболее перспективного кандидата

### Согласно литературным данным, взаимоотношения в семье влияют на зависимость от смартфона: чем хуже отношения с родителями, тем сильнее зависимость от смартфона. также билингвы обычно обладают более развитым исполнительным контролем — то есть способностью управлять вниманием, подавлять отвлекающие стимулы и переключаться между задачами. 

Поэтому: 
1) они могут лучше регулировать импульсивное поведение (например, желание проверить уведомления). 
2) У них выше устойчивость внимания и самоконтроль, что снижает риск формирования зависимости от смартфона.

SNP 1 - входит в состав 7 экзона гена COMT - кодирующего катехол-о-метилтрансферазу, которая метаболизирует катехоломины. SNP влияет на активность гена COMT. G - высокая активность COMT, A - низкая.

0 - GG
1 - AG
2 - AA

Анализ полиморфизма rs4680 с учетом этнической принадлежности показал значимую ассоциацию у башкир: носители генотипа AA демонстрировали более высокую склонность к смартфон-зависимости по сравнению с другими группами. В то же время у русских и татар статистически значимых различий обнаружено не было.

Биологически это можно объяснить тем, что генотип AA (Met/Met) снижает активность COMT, повышая уровень дофамина, что усиливает импульсивность и поиск вознаграждения, включая чрезмерное использование смартфона. Однако эффект локуса rs4680 слишком слаб, чтобы проявляться сам по себе, и становится заметным только в сочетании с другими генетическими или средовыми факторами, которые могут различаться у этнических групп.
