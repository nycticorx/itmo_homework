## Загрузка библиотек



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Загрузка данных

```python
df = pd.read_parquet("my_project.parquet")
```

# Первичный обзор

```python
print("Shape:", df.shape)
df.info()   # Чтобы понять структуру и полноту данных
display(df.describe(include='all').T) 
completeness = df.notnull().mean() * 100  #заполненность данных в процентах
completeness = completeness.sort_values(ascending=False)  # чтобы отсортировать
completeness

duplicates = df.duplicated().sum()
print(f"Количество полных дубликатов: {duplicates}")
unique_ids = df['id'].is_unique
print(f"Столбец 'id' содержит только уникальные значения? {unique_ids}")
df[df.duplicated()]
```

### **Выводы о данных**

- Датасет содержит 376 наблюдений и 51 переменную, включая демографические, медицинские и генетические показатели.
Присутствуют числовые, категориальные и буллевые признаки.

 - Большинство демографических и генетических переменных заполнено более чем на 90%, однако некоторые столбцы (Family_amount, Fin_level, SNP2_78171124) содержат менее 30 % данных. Эти признаки требуют проверки и исключения из анализа на следующих этапах.

- В переменной a_full_fledged_family обнаружен выброс (значение 167 при типичном диапазоне 1–2). Требуется проверка или корректировка данных.

- В выборке большинство участников не курят, имеют татарскую национальность и владеют двумя языками. Категориальные данные выглядят логично, явных аномалий нет. Преобладают женщины. 

- Большинство SNP-переменных варьируются в диапазоне 0–2 (генотипы), что указывает на корректное кодирование. SNP2_78171124 не несёт вариативности и может быть исключён из анализа.

- Проверка структуры данных показала, что в датасете отсутствуют полные дубликаты строк.
- Столбец id содержит только уникальные значения, что подтверждает корректную идентификацию наблюдений.
